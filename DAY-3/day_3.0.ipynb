```python
height = float(input("enter your height in m: "))
weight = float(input("enter your weight in kg: "))

BmI = weight/height**2

typeshi = ""
if BmI<=18.5:
    typeshi = "Underweight"
elif BmI<=25:
    typeshi = "Normal-Weight"
elif BmI<=30:
    typeshi = "Over-weight"
else:
    typeshi = "Clinically Obese"
print(f"Your bim is {round(BmI)} and you are {typeshi}.")
```

<p style='text-align:center';><b>LEaP YeaR

```python
user_year = int(input("Which year do you want to check? "))
leap_year = True
if user_year%4==0:
    leap_year = True
    if user_year%100==0:
        leap_year = False
        if user_year%400==0:
            leap_year = True
        else:
            leap_year = False
    else:
        leap_year = True
else:
    leap_year = False

if leap_year == True:

    print(f"The year {user_year} is a leap year")
else:
    print(f"The year {user_year} is NOT a leap year")
```
