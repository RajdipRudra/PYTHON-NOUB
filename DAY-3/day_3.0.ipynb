```python
user_year = int(input("Which year do you want to check? "))
leap_year = True
if user_year%4==0:
    leap_year = True
    if user_year%100==0:
        leap_year = False
        if user_year%400==0:
            leap_year = True
        else:
            leap_year = False
    else:
        leap_year = True
else:
    leap_year = False

if leap_year == True:

    print(f"The year {user_year} is a leap year")
else:
    print(f"The year {user_year} is NOT a leap year")
```
