<h5>Swap the value of 2 variables a & b

```python
#Easy and besic one ->
a = 10
b = 40

print(f"a = {a} before")
print(f"b = {b} before")

c=a
a=b
b=c

print("\n")

print(f"a = {a} after")
print(f"b = {b} after")
```

```python
#Hard but not That hard one ->

a = 29
b = 21

print(f"a = {a} before")
print(f"b = {b} before")

a = a+b
b = a-b
a = a-b

print("\n")

print(f"a = {a} after")
print(f"b = {b} after")
```

```python
#well what if the values ain't even number then what?🤔🤔🤔 ->

#welllllll...Getting the ASCII value in Python is straightforward using the built-in { ord() } function.

a = 'Q'
b = 'R'

print(f"a = {a} before")
print(f"b = {b} before")

a = ord(a)
b = ord(b)


a = a+b
b = a-b
a = a-b


a = chr(a)
b = chr(b)

print("\n")

print(f"a = {a} after")
print(f"b = {b} after")
```

```python
### well then what about strings? welll ->

a = "Rajdip"
b = "JOyDip"

print(f"a = {a} before")
print(f"b = {b} before")

unity = []
if (len(a) > 1) or (len(b) > 1):
    A = []
    for i in a:
        A.append(ord(i))
        unity.append(ord(i))

    B = []
    for j in b:
        B.append(ord(j))
        unity.append(ord(j))
    
    Aa = []
    Bb = []
    # print(unity)
    for i in range(round(len(unity)/2)):
        num1 = A[i]
        num2 = B[i]

        num1 = num1+num2
        num2 = num1-num2
        num1 = num1-num2
        
        Aa.append(num1)
        Bb.append(num2)


a = []
for l in Aa:
    a.append(chr(l))

b = []
for j in Bb:
    b.append(chr(j))

print("\n")
# print(A)
# print(B)
print(f"a = {"".join(a)} after")
print(f"b = {"".join(b)} after")
```
